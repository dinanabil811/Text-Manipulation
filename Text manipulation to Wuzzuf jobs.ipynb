# Convert the first char (only) in the sentence to capital and remove spaces on the left and right

```python
for i in range(len(data["Company"])):
    data['Job Title'][i] = str.title(data['Job Title'][i])
    data['Job Title'][i].strip()
    data['State'][i] = str.title(data['State'][i])
    data['State'][i].strip()
    data['Company'][i] = str.title(data['Company'][i])
    data['Company'][i].strip()
    data['Description'][i] = str.title(data['Description'][i])
    data['Description'][i].strip()
    data['Governorate'][i] = str.title(data['Governorate'][i])
    data['Governorate'][i].strip()
```
